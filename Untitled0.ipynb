```python
# Импорт библиотек
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from collections import defaultdict
import warnings
warnings.filterwarnings('ignore')

# Настройка отображения графиков в Colab
%matplotlib inline
```

```python
# Классы Elo и Glicko

class EloSystem:
    """
    Реализация рейтинговой системы Эло.
    """
    def __init__(self, k=32, initial_rating=1500, scale=400, base=10):
        self.k = k
        self.initial_rating = initial_rating
        self.scale = scale
        self.base = base
        self.ratings = {}          # player_id -> текущий рейтинг

    def add_player(self, player_id, rating=None):
        if rating is None:
            rating = self.initial_rating
        self.ratings[player_id] = rating

    def expected_score(self, rating_a, rating_b):
        """Вероятность победы игрока A над B."""
        return 1.0 / (1.0 + self.base ** ((rating_b - rating_a) / self.scale))

    def predict_match(self, player_a, player_b):
        """Вероятность победы player_a (ожидаемый счёт)."""
        r_a = self.ratings.get(player_a, self.initial_rating)
        r_b = self.ratings.get(player_b, self.initial_rating)
        return self.expected_score(r_a, r_b)

    def update(self, player_a, player_b, outcome_a):
        """
        Обновляет рейтинги после матча.
        outcome_a: 1.0 - победа player_a, 0.0 - победа player_b, 0.5 - ничья.
        """
        if player_a not in self.ratings:
            self.add_player(player_a)
        if player_b not in self.ratings:
            self.add_player(player_b)

        r_a = self.ratings[player_a]
        r_b = self.ratings[player_b]
        e_a = self.expected_score(r_a, r_b)
        delta = self.k * (outcome_a - e_a)
        self.ratings[player_a] += delta
        self.ratings[player_b] -= delta


class GlickoSystem:
    """
    Реализация рейтинговой системы Glicko (версия 1).
    """
    def __init__(self, c=34.6, initial_rating=1500, initial_rd=350,
                 min_rd=30, max_rd=350):
        self.c = c
        self.initial_rating = initial_rating
        self.initial_rd = initial_rd
        self.min_rd = min_rd
        self.max_rd = max_rd
        self.ratings = {}          # player_id -> рейтинг
        self.rds = {}              # player_id -> рейтинговое отклонение
        self.last_round = {}       # player_id -> последний сыгранный тур
        self.q = math.log(10) / 400.0  # константа

    def add_player(self, player_id, rating=None, rd=None, round_num=0):
        if rating is None:
            rating = self.initial_rating
        if rd is None:
            rd = self.initial_rd
        self.ratings[player_id] = rating
        self.rds[player_id] = rd
        self.last_round[player_id] = round_num

    def _increase_rd(self, player_id, current_round):
        """Увеличивает RD за счёт пропущенных туров."""
        if player_id not in self.last_round:
            self.add_player(player_id, round_num=current_round)
            return
        t = current_round - self.last_round[player_id]
        if t > 0:
            rd = self.rds[player_id]
            new_rd = math.sqrt(rd**2 + (self.c**2) * t)
            new_rd = min(new_rd, self.max_rd)
            self.rds[player_id] = new_rd
            self.last_round[player_id] = current_round

    def g(self, rd):
        """Вспомогательная функция g(RD)."""
        return 1.0 / math.sqrt(1.0 + 3.0 * (self.q**2) * (rd**2) / (math.pi**2))

    def expected_score(self, player_a, player_b):
        """Вероятность победы player_a с учётом RD противника."""
        r_a = self.ratings.get(player_a, self.initial_rating)
        r_b = self.ratings.get(player_b, self.initial_rating)
        rd_b = self.rds.get(player_b, self.initial_rd)
        g_b = self.g(rd_b)
        return 1.0 / (1.0 + 10.0 ** (-g_b * (r_a - r_b) / 400.0))

    def predict_match(self, player_a, player_b):
        """Вероятность победы player_a (ожидаемый счёт)."""
        return self.expected_score(player_a, player_b)

    def update(self, player_a, player_b, outcome_a, round_num):
        """
        Обновляет рейтинги и RD после матча.
        outcome_a: 1.0 - победа player_a, 0.0 - победа player_b, 0.5 - ничья.
        """
        if player_a not in self.ratings:
            self.add_player(player_a, round_num=round_num)
        if player_b not in self.ratings:
            self.add_player(player_b, round_num=round_num)

        self._increase_rd(player_a, round_num)
        self._increase_rd(player_b, round_num)

        r_a = self.ratings[player_a]
        r_b = self.ratings[player_b]
        rd_a = self.rds[player_a]
        rd_b = self.rds[player_b]

        g_b = self.g(rd_b)
        e_a = 1.0 / (1.0 + 10.0 ** (-g_b * (r_a - r_b) / 400.0))

        d2_inv = (self.q**2) * (g_b**2) * e_a * (1.0 - e_a)
        if d2_inv == 0:
            d2 = float('inf')
        else:
            d2 = 1.0 / d2_inv

        if d2 != float('inf'):
            delta_r = (self.q / (1.0 / (rd_a**2) + 1.0 / d2)) * g_b * (outcome_a - e_a)
            self.ratings[player_a] = r_a + delta_r
            new_rd = math.sqrt(1.0 / (1.0 / (rd_a**2) + 1.0 / d2))
            self.rds[player_a] = max(new_rd, self.min_rd)

        g_a = self.g(rd_a)
        e_b = 1.0 - e_a
        outcome_b = 1.0 - outcome_a
        d2_inv_b = (self.q**2) * (g_a**2) * e_b * (1.0 - e_b)
        if d2_inv_b == 0:
            d2_b = float('inf')
        else:
            d2_b = 1.0 / d2_inv_b

        if d2_b != float('inf'):
            delta_r_b = (self.q / (1.0 / (rd_b**2) + 1.0 / d2_b)) * g_a * (outcome_b - e_b)
            self.ratings[player_b] = r_b + delta_r_b
            new_rd_b = math.sqrt(1.0 / (1.0 / (rd_b**2) + 1.0 / d2_b))
            self.rds[player_b] = max(new_rd_b, self.min_rd)

        self.last_round[player_a] = round_num
        self.last_round[player_b] = round_num
```

```python
# Вспомогательные функции

def run_simulation(matches_df, system):
    """
    Запускает симуляцию по данным матчей.
    system: объект EloSystem или GlickoSystem.
    Возвращает:
    - history_ratings: список словарей {player: rating} после каждого тура (включая начальное состояние).
    - history_rds: (только для Glicko) список словарей {player: rd} после каждого тура.
    - predictions: список кортежей (player_a, player_b, actual_outcome, predicted_prob).
    """
    all_players = set(matches_df['player1']).union(set(matches_df['player2']))
    for p in all_players:
        if p not in system.ratings:
            if isinstance(system, GlickoSystem):
                system.add_player(p, round_num=0)
            else:
                system.add_player(p)

    rounds = sorted(matches_df['round'].unique())
    history_ratings = []
    history_rds = [] if isinstance(system, GlickoSystem) else None
    predictions = []

    history_ratings.append(system.ratings.copy())
    if history_rds is not None:
        history_rds.append(system.rds.copy())

    for rnd in rounds:
        matches_round = matches_df[matches_df['round'] == rnd]
        for _, row in matches_round.iterrows():
            p1 = row['player1']
            p2 = row['player2']
            outcome = row['outcome']

            pred = system.predict_match(p1, p2)
            predictions.append((p1, p2, outcome, pred))

            if isinstance(system, GlickoSystem):
                system.update(p1, p2, outcome, rnd)
            else:
                system.update(p1, p2, outcome)

        history_ratings.append(system.ratings.copy())
        if history_rds is not None:
            history_rds.append(system.rds.copy())

    return history_ratings, history_rds, predictions


def evaluate_predictions(predictions, ignore_draws=True):
    """
    Вычисляет метрики качества предсказаний.
    predictions: список (player_a, player_b, actual_outcome, pred_prob)
    ignore_draws: если True, то матчи с ничьей (actual=0.5) не учитываются при расчёте accuracy.
    Возвращает словарь с метриками: accuracy, log_loss, brier_score.
    """
    actuals = []
    preds = []
    for _, _, actual, pred in predictions:
        if ignore_draws and actual == 0.5:
            continue
        actuals.append(actual)
        preds.append(pred)

    if len(actuals) == 0:
        return {'accuracy': np.nan, 'log_loss': np.nan, 'brier_score': np.nan}

    actuals = np.array(actuals)
    preds = np.array(preds)

    if ignore_draws:
        pred_classes = (preds >= 0.5).astype(int)
        actual_classes = actuals.astype(int)
        accuracy = np.mean(pred_classes == actual_classes)
    else:
        accuracy = np.nan

    mask = (actuals != 0.5) if ignore_draws else np.ones_like(actuals, dtype=bool)
    if np.sum(mask) == 0:
        log_loss = np.nan
        brier = np.nan
    else:
        actuals_bin = actuals[mask]
        preds_bin = preds[mask]
        eps = 1e-15
        preds_bin = np.clip(preds_bin, eps, 1 - eps)
        log_loss = -np.mean(actuals_bin * np.log(preds_bin) + (1 - actuals_bin) * np.log(1 - preds_bin))
        brier = np.mean((preds_bin - actuals_bin) ** 2)

    return {
        'accuracy': accuracy,
        'log_loss': log_loss,
        'brier_score': brier
    }


def plot_top_players(history_ratings, top_n=10, title="Rating over Rounds", save_path=None):
    """Строит график изменения рейтинга топ-N игроков."""
    if not history_ratings:
        return
    final_ratings = history_ratings[-1]
    top_players = sorted(final_ratings.items(), key=lambda x: x[1], reverse=True)[:top_n]
    top_ids = [p for p, _ in top_players]

    rounds = list(range(len(history_ratings)))
    plt.figure(figsize=(12, 6))
    for player in top_ids:
        ratings_over_time = [state.get(player, np.nan) for state in history_ratings]
        plt.plot(rounds, ratings_over_time, marker='o', label=player)

    plt.xlabel("Round")
    plt.ylabel("Rating")
    plt.title(title)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()


def plot_rd_history(history_rds, players=None, title="RD over Rounds", save_path=None):
    """Строит график изменения RD для заданных игроков."""
    if history_rds is None or not history_rds:
        return
    if players is None:
        all_players = list(history_rds[0].keys())
        players = all_players[:5]

    rounds = list(range(len(history_rds)))
    plt.figure(figsize=(12, 6))
    for player in players:
        rd_over_time = [state.get(player, np.nan) for state in history_rds]
        plt.plot(rounds, rd_over_time, marker='s', label=player)

    plt.xlabel("Round")
    plt.ylabel("Rating Deviation (RD)")
    plt.title(title)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()


def plot_accuracy_comparison(preds_elo, preds_glicko, window=5, save_path=None):
    """Строит график скользящей точности для двух моделей."""
    def compute_rolling_accuracy(predictions, window):
        actuals = [act for _, _, act, _ in predictions if act != 0.5]
        preds = [pred for _, _, act, pred in predictions if act != 0.5]
        if len(actuals) == 0:
            return []
        pred_classes = [1 if p >= 0.5 else 0 for p in preds]
        correct = [1 if pc == ac else 0 for pc, ac in zip(pred_classes, actuals)]
        rolling_acc = []
        for i in range(len(correct)):
            start = max(0, i - window + 1)
            rolling_acc.append(np.mean(correct[start:i+1]))
        return rolling_acc

    acc_elo = compute_rolling_accuracy(preds_elo, window)
    acc_glicko = compute_rolling_accuracy(preds_glicko, window)

    if not acc_elo or not acc_glicko:
        print("Недостаточно данных для построения точности.")
        return

    plt.figure(figsize=(12, 6))
    plt.plot(range(len(acc_elo)), acc_elo, label='Elo', marker='.')
    plt.plot(range(len(acc_glicko)), acc_glicko, label='Glicko', marker='.')
    plt.xlabel("Match index (rolling window)")
    plt.ylabel("Accuracy")
    plt.title(f"Rolling Accuracy Comparison (window={window})")
    plt.legend()
    plt.grid(True)
    plt.ylim([0, 1])
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
# Загрузка и подготовка данных из games.csv

# Путь к вашему файлу на Google Диске
file_path = '/content/drive/MyDrive/games.csv'

# Загрузка данных
df_raw = pd.read_csv(file_path)
print(f"Загружено {len(df_raw)} записей.")
print("Первые 5 строк сырых данных:")
print(df_raw.head())

# ------------------ Преобразование данных ------------------
# Сортируем по времени создания игры (чтобы туры шли в хронологическом порядке)
df_raw = df_raw.sort_values('created_at').reset_index(drop=True)

# Создаём колонку outcome: 1.0 - победа white, 0.0 - победа black, 0.5 - ничья
def get_outcome(row):
    if row['winner'] == 'white':
        return 1.0
    elif row['winner'] == 'black':
        return 0.0
    else:  # 'draw'
        return 0.5

df_raw['outcome'] = df_raw.apply(get_outcome, axis=1)

# Задаём туры: группируем по 50 матчей в одном туре
# (вы можете изменить это число в зависимости от плотности игр)
MATCHES_PER_ROUND = 50
df_raw['round'] = (df_raw.index // MATCHES_PER_ROUND) + 1

# Переименовываем колонки в единый формат
df = df_raw.rename(columns={'white_id': 'player1', 'black_id': 'player2'})

# Оставляем только нужные колонки
df = df[['round', 'player1', 'player2', 'outcome']]

# Фильтруем игроков, у которых мало игр (оставляем только сыгравших >= 5 партий)
player_counts = pd.concat([df['player1'], df['player2']]).value_counts()
active_players = player_counts[player_counts >= 5].index.tolist()
df = df[df['player1'].isin(active_players) & df['player2'].isin(active_players)]

print(f"\nПосле фильтрации осталось {len(df)} записей.")
print("Первые 5 строк подготовленных данных:")
print(df.head())
print(f"\nКоличество туров: {df['round'].nunique()}")
print(f"Количество игроков: {len(set(df['player1']).union(set(df['player2'])))}")
```

```python
# Запуск симуляции

# Инициализация систем
elo = EloSystem(k=32)
glicko = GlickoSystem(c=34.6)

print("Запуск симуляции Elo...")
elo_history, _, elo_preds = run_simulation(df, elo)
print(f"Симуляция Elo завершена. Обработано {len(elo_preds)} матчей.")

print("Запуск симуляции Glicko...")
glicko_history, glicko_rds_history, glicko_preds = run_simulation(df, glicko)
print(f"Симуляция Glicko завершена. Обработано {len(glicko_preds)} матчей.")
```

```python
# Оценка точности моделей

metrics_elo = evaluate_predictions(elo_preds, ignore_draws=True)
metrics_glicko = evaluate_predictions(glicko_preds, ignore_draws=True)

# Создаём DataFrame для красивого вывода
metrics_df = pd.DataFrame({
    'Metric': ['Accuracy', 'Log Loss', 'Brier Score'],
    'Elo': [metrics_elo['accuracy'], metrics_elo['log_loss'], metrics_elo['brier_score']],
    'Glicko': [metrics_glicko['accuracy'], metrics_glicko['log_loss'], metrics_glicko['brier_score']]
})

print("\n=== Сравнение метрик (без учёта ничьих) ===")
print(metrics_df.to_string(index=False))
```

```python
# Визуализация

# 1. Топ-10 по Elo
plot_top_players(elo_history, top_n=10, title="Elo Rating over Rounds (Top 10)")

# 2. Топ-10 по Glicko
plot_top_players(glicko_history, top_n=10, title="Glicko Rating over Rounds (Top 10)")

# 3. График RD для первых 5 игроков (по алфавиту или первым в списке)
# Для наглядности возьмём первых 5 игроков из истории
sample_players = list(glicko_rds_history[0].keys())[:5]
plot_rd_history(glicko_rds_history, players=sample_players, title="Glicko RD over Rounds (Sample Players)")

# 4. Сравнение точности (скользящее окно)
plot_accuracy_comparison(elo_preds, glicko_preds, window=20)
```

```python
# Сохранение результатов

# Собираем итоговые рейтинги в один DataFrame
final_elo = pd.DataFrame(list(elo.ratings.items()), columns=['Player', 'Elo_Rating'])
final_glicko = pd.DataFrame(list(glicko.ratings.items()), columns=['Player', 'Glicko_Rating'])
final_rd = pd.DataFrame(list(glicko.rds.items()), columns=['Player', 'RD'])

final_df = final_elo.merge(final_glicko, on='Player').merge(final_rd, on='Player')
final_df = final_df.sort_values('Elo_Rating', ascending=False)

# Сохраняем на Google Диск
output_path = '/content/drive/MyDrive/final_ratings.csv'
final_df.to_csv(output_path, index=False)
print(f"Итоговые рейтинги сохранены в {output_path}")
print(final_df.head(10))
```
